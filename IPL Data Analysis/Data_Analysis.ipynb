```
  
# nesting aggregate function  MAX and Sum.


spark.sql("""
SELECT  s.bowling_team, max(s.SumExRun) AS extra_runs_in_match
FROM (SELECT match_id, bowling_team, SUM(extra_runs) SumExRun
      FROM ipl_ball_by_ball_csv 
      GROUP BY match_id, bowling_team) s
GROUP BY bowling_team
ORDER BY max(s.SumExRun) DESC """).display()

#note:- 
#instead of nesting aggregate function here we can  untilize backend  MAX aggregate using 
#display() and order by SUM(extra_run) in Pivot table or bar graph this can eradicate performance plenty.
         
```

```
#the Leading wicket taker in the IPL. - SL Malinga|         170
#select bowler and dismissal_kind coloums and use where filter out if the dismissal is run out,obstructing the field and wicket is not taken.

spark.sql("""
SELECT bowler,COUNT(dismissal_kind) AS wicket_taken
FROM ipl_ball_by_ball_csv
WHERE (dismissal_kind != 'NA' AND dismissal_kind != 'obstructing the field' AND dismissal_kind != 'run out')
GROUP BY bowler
ORDER BY wicket_taken DESC

""").display()
```

```
# joining two ipl_matches_csv and table ipl_ball_by_ball_csv using NATURAL JOIN as we only need one common columns and  condition it to only D/L method  
#since we need higher run scorer we use group by matches and batsman.

spark.sql("""
SELECT match_id,sum(total_runs),batsman
FROM  ipl_matches_csv
NATURAL JOIN ipl_ball_by_ball_csv
where (method = 'D/L')
group by match_id,batsman
order by sum(total_runs) DESC
""").show(1)
```

```
#remember don't use overs in group by use only if you need strike rate for each over
spark.sql("""
SELECT batsman,(sum(batsman_runs)/count(batsman_runs))*100 AS strike_rate
FROM ipl_ball_by_ball_csv
WHERE overs BETWEEN 1 AND 6 
GROUP BY batsman
ORDER BY strike_rate DESC
""").display()
```

```
#given to 10 players so group by batsman
#since a batman can also count as out so append  batsman position and non_strike to count all the dismissal. 
#short the querry according to batting_average in descending order.
spark.sql("""
SELECT batsman AS player, SUM(batsman_runs)/COUNT(player_dismissed) AS batting_average
FROM
(
       SELECT batsman, batsman_runs, player_dismissed 
       FROM ipl_ball_by_ball_csv
       UNION ALL 
       SELECT non_striker, batsman_runs, player_dismissed
       FROM ipl_ball_by_ball_csv
)
GROUP BY player
ORDER BY batting_average DESC""").show(10,truncate =False)
```

```
#appending both umpire columns in umpire1 using UNION ALL and the count all  the no. of time each umpire present.

spark.sql(""" 
SELECT umpire1 AS umpires ,COUNT(umpire1) total_officiated
FROM
(
    SELECT umpire1
    FROM ipl_matches_csv
    UNION ALL
    SELECT umpire2
    FROM ipl_matches_csv
) 
GROUP BY umpires
ORDER BY total_officiated DESC
""").show(1,truncate =False)
```
